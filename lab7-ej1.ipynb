### Introducción

El Campeonato Mundial de la FIFA 2026 ha popularizado el álbum Panini con **980 estampas distintas**, 7 por sobre. En esta primera etapa usamos un álbum reducido de **N = 100** estampas para familiarizarnos con la simulación y compararla con la **teoría del coleccionista** (*Coupon Collector Problem*).

La teoría establece que el número esperado de sobres para completar el álbum es:

$$E[T] \approx \frac{N}{S} \cdot H_N, \quad H_N = \sum_{k=1}^{N} \frac{1}{k} \approx \ln(N) + \gamma$$

donde $\gamma \approx 0.5772$ es la constante de Euler-Mascheroni.

```python
import numpy as np
import matplotlib.pyplot as plt
import math

plt.rcParams.update({'figure.dpi': 110, 'font.size': 11})
```

## 1. Parámetros de la simulación

```python
N    = 100    # número total de estampas distintas
S    = 7      # estampas por sobre (todas distintas dentro del sobre)
R    = 10_000 # número de simulaciones
SEED = 2026   # semilla obligatoria para reproducibilidad

print(f"N={N} | S={S} | R={R:,} | semilla={SEED}")
```

## 2. Simulación de Monte Carlo

Procedimiento por simulación:
1. Inicializar arreglo booleano `album` (False = no obtenida).
2. Inicializar contadores de sobres y repetidas en cero.
3. Comprar sobres hasta tener las N estampas distintas.
4. Almacenar resultados.

```python
np.random.seed(SEED)

sobres_result    = np.zeros(R, dtype=int)
repetidas_result = np.zeros(R, dtype=int)

for i in range(R):
    album     = np.zeros(N, dtype=bool)
    sobres    = 0
    repetidas = 0
    count     = 0  # estampas distintas obtenidas

    while count < N:
        stickers  = np.random.choice(N, size=S, replace=False)
        nuevas    = ~album[stickers]          # máscara de estampas nuevas
        repetidas += S - int(nuevas.sum())    # las que ya teníamos
        count     += int(nuevas.sum())
        album[stickers] = True
        sobres += 1

    sobres_result[i]    = sobres
    repetidas_result[i] = repetidas

print(f"Simulación completada: {R:,} repeticiones.")
```

## 3. Resultados estadísticos

```python
media_sobres    = np.mean(sobres_result)
std_sobres      = np.std(sobres_result)
media_rep       = np.mean(repetidas_result)
std_rep         = np.std(repetidas_result)
minimo_teorico  = math.ceil(N / S)   # mínimo posible = ⌈100/7⌉ = 15
prob_mas_30     = np.mean(sobres_result > 30)

sep = "=" * 58
print(sep)
print(f"  RESULTADOS  (N={N}, S={S}, R={R:,}, semilla={SEED})")
print(sep)
print(f"\n  Sobres necesarios para completar el álbum:")
print(f"    Media          : {media_sobres:.4f}")
print(f"    Desv. estándar : {std_sobres:.4f}")
print(f"\n  Estampas repetidas acumuladas:")
print(f"    Media          : {media_rep:.4f}")
print(f"    Desv. estándar : {std_rep:.4f}")
print(f"\n  Mínimo teórico de sobres : {minimo_teorico}  (⌈{N}/{S}⌉)")
print(f"  P(sobres > 30)           : {prob_mas_30:.4f}  ({prob_mas_30*100:.2f}%)")
print(sep)
```

> **Justificación del umbral 30:** El mínimo teórico (sin ninguna repetida) es $\lceil 100/7 \rceil = 15$ sobres. El umbral de 30 sobres es el doble de ese mínimo, representando un escenario con algo de repetición pero aún muy favorable. Sirve para cuantificar qué tan probable es completar el álbum con pocos sobres.
